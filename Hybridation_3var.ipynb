```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('rn-as+sem+soc_75.txt', delimiter='\t')

print(df.shape)
df.columns = ['alpha', 'beta','gamma','mae','rmse'] 


print(df.head)

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')



x = df[['alpha']]
y = df[['beta']]
z = df[['gamma']]

c = df[['mae']]

print(df[df.rmse == df.rmse.min()]) 
print(df[df.mae == df.mae.min()]) 

ax.set_xlabel("alpha")
ax.set_ylabel("beta")
ax.set_zlabel("gamma")

img = ax.scatter(x, y, z, c=c, cmap=plt.hot(), depthshade=True)


ax.view_init(30, 40)
fig.colorbar(img)
plt.show()

fig.savefig("MAExyz")
```
